```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.tsa.seasonal import seasonal_decompose

# Đọc dữ liệu
df = pd.read_csv('bt.csv')

# Kiểm tra dữ liệu
df
```

```python


# Visualize the target variables
plt.figure(figsize=(10, 6))
plt.plot(df['6'], label='6')
plt.plot(df['7'], label='7')
plt.title('Time Series of Target Variables')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()

# Decompose the time series for both variables
decomposition_6 = seasonal_decompose(df['6'], period=12, model='additive')
fig_6 = decomposition_6.plot()
plt.title('Decomposition of Variable 6')
plt.show()

decomposition_7 = seasonal_decompose(df['7'], period=12, model='additive')
fig_7 = decomposition_7.plot()
plt.title('Decomposition of Variable 7')
plt.show()

# Plot the autocorrelation and partial autocorrelation for both variables
fig, ax = plt.subplots(4, 1, figsize=(12, 16))

sm.graphics.tsa.plot_acf(df['6'], lags=50, ax=ax[0])
ax[0].set_title('Autocorrelation of Variable 6')

sm.graphics.tsa.plot_pacf(df['6'], lags=50, ax=ax[1])
ax[1].set_title('Partial Autocorrelation of Variable 6')

sm.graphics.tsa.plot_acf(df['7'], lags=50, ax=ax[2])
ax[2].set_title('Autocorrelation of Variable 7')

sm.graphics.tsa.plot_pacf(df['7'], lags=50, ax=ax[3])
ax[3].set_title('Partial Autocorrelation of Variable 7')

plt.tight_layout()
plt.show()

# Plot rolling statistics for both variables
rolling_mean_6 = df['6'].rolling(window=12).mean()
rolling_std_6 = df['6'].rolling(window=12).std()

rolling_mean_7 = df['7'].rolling(window=12).mean()
rolling_std_7 = df['7'].rolling(window=12).std()

plt.figure(figsize=(12, 8))
plt.plot(df['6'], label='Variable 6')
plt.plot(rolling_mean_6, label='Rolling Mean (Variable 6)')
plt.plot(rolling_std_6, label='Rolling Std (Variable 6)')

plt.plot(df['7'], label='Variable 7')
plt.plot(rolling_mean_7, label='Rolling Mean (Variable 7)')
plt.plot(rolling_std_7, label='Rolling Std (Variable 7)')

plt.title('Rolling Mean & Standard Deviation')
plt.legend()
plt.show()

# Check for stationarity using the Augmented Dickey-Fuller test for both variables
from statsmodels.tsa.stattools import adfuller

result_6 = adfuller(df['6'])
print('ADF Statistic (Variable 6): %f' % result_6[0])
print('p-value (Variable 6): %f' % result_6[1])
print('Critical Values (Variable 6):')
for key, value in result_6[4].items():
    print('\t%s: %.3f' % (key, value))

result_7 = adfuller(df['7'])
print('\nADF Statistic (Variable 7): %f' % result_7[0])
print('p-value (Variable 7): %f' % result_7[1])
print('Critical Values (Variable 7):')
for key, value in result_7[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C
from sklearn.metrics import mean_squared_error

# Đọc dữ liệu
df = pd.read_csv('bt.csv')

# Chuẩn bị dữ liệu cho cột '6'
X_6 = np.arange(len(df)).reshape(-1, 1)
y_6 = df['6'].values

# Chuẩn bị dữ liệu cho cột '7'
X_7 = np.arange(len(df)).reshape(-1, 1)
y_7 = df['7'].values

# Chia dữ liệu thành tập huấn luyện và kiểm tra
split = int(0.8 * len(df))

# Chia cho cột '6'
X_train_6, X_test_6 = X_6[:split], X_6[split:]
y_train_6, y_test_6 = y_6[:split], y_6[split:]

# Chia cho cột '7'
X_train_7, X_test_7 = X_7[:split], X_7[split:]
y_train_7, y_test_7 = y_7[:split], y_7[split:]

# Xây dựng kernel cho Gaussian Process
kernel = C(1.0, (1e-3, 1e3)) * RBF(1.0, (1e-2, 1e2))

# Gaussian Process Regression cho cột '6'
gp_6 = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10, normalize_y=True)
gp_6.fit(X_train_6, y_train_6)
y_pred_6, sigma_6 = gp_6.predict(X_test_6, return_std=True)

# Vẽ kết quả cho cột '6'
plt.figure(figsize=(10, 6))
plt.plot(X_6, y_6, 'k.', markersize=10, label='Observations')
plt.plot(X_test_6, y_pred_6, 'r-', label='Prediction')
plt.fill_between(X_test_6.ravel(), y_pred_6 - 1.96 * sigma_6, y_pred_6 + 1.96 * sigma_6, alpha=0.2, label='95% confidence interval')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Gaussian Process Regression for Column 6')
plt.legend()
plt.show()



# Gaussian Process Regression cho cột '7'
gp_7 = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10, normalize_y=True)
gp_7.fit(X_train_7, y_train_7)
y_pred_7, sigma_7 = gp_7.predict(X_test_7, return_std=True)

# Vẽ kết quả cho cột '7'
plt.figure(figsize=(10, 6))
plt.plot(X_7, y_7, 'k.', markersize=10, label='Observations')
plt.plot(X_test_7, y_pred_7, 'r-', label='Prediction')
plt.fill_between(X_test_7.ravel(), y_pred_7 - 1.96 * sigma_7, y_pred_7 + 1.96 * sigma_7, alpha=0.2, label='95% confidence interval')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Gaussian Process Regression for Column 7')
plt.legend()
plt.show()

from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, r2_score

# Mean Absolute Error (MAE)
mae_6 = mean_absolute_error(y_test_6, y_pred_6)
mae_7 = mean_absolute_error(y_test_7, y_pred_7)

print('Mean Absolute Error (Column 6):', mae_6)
print('Mean Absolute Error (Column 7):', mae_7)

# Root Mean Squared Error (RMSE)
rmse_6 = mean_squared_error(y_test_6, y_pred_6, squared=False)
rmse_7 = mean_squared_error(y_test_7, y_pred_7, squared=False)

print('Root Mean Squared Error (Column 6):', rmse_6)
print('Root Mean Squared Error (Column 7):', rmse_7)

# Mean Absolute Percentage Error (MAPE)
def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

mape_6 = mean_absolute_percentage_error(y_test_6, y_pred_6)
mape_7 = mean_absolute_percentage_error(y_test_7, y_pred_7)

print('Mean Absolute Percentage Error (Column 6):', mape_6)
print('Mean Absolute Percentage Error (Column 7):', mape_7)

# R-squared (R2)
r2_6 = r2_score(y_test_6, y_pred_6)
r2_7 = r2_score(y_test_7, y_pred_7)

print('R-squared (Column 6):', r2_6)
print('R-squared (Column 7):', r2_7)

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller
from sklearn.metrics import mean_squared_error



# Kiểm tra tính dừng cho cột '6'
result_6 = adfuller(df['6'])
print('ADF Statistic (Column 6): %f' % result_6[0])
print('p-value (Column 6): %f' % result_6[1])
print('Critical Values (Column 6):')
for key, value in result_6[4].items():
    print('\t%s: %.3f' % (key, value))

# Kiểm tra tính dừng cho cột '7'
result_7 = adfuller(df['7'])
print('\nADF Statistic (Column 7): %f' % result_7[0])
print('p-value (Column 7): %f' % result_7[1])
print('Critical Values (Column 7):')
for key, value in result_7[4].items():
    print('\t%s: %.3f' % (key, value))

# Chia dữ liệu thành tập huấn luyện và kiểm tra
split = int(0.8 * len(df))
train_6, test_6 = df['6'][:split], df['6'][split:]
train_7, test_7 = df['7'][:split], df['7'][split:]

# Xây dựng mô hình SARIMA cho cột '6'
model_6 = SARIMAX(train_6, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
model_fit_6 = model_6.fit(disp=False)

# Dự báo cho cột '6'
y_pred_6 = model_fit_6.predict(start=len(train_6), end=len(train_6) + len(test_6) - 1, dynamic=False)

# Vẽ kết quả cho cột '6'
plt.figure(figsize=(10, 6))
plt.plot(train_6.index, train_6, label='Train (Column 6)')
plt.plot(test_6.index, test_6, label='Test (Column 6)')
plt.plot(test_6.index, y_pred_6, label='SARIMA Prediction (Column 6)')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('SARIMA Model for Column 6')
plt.legend()
plt.show()



# Xây dựng mô hình SARIMA cho cột '7'
model_7 = SARIMAX(train_7, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
model_fit_7 = model_7.fit(disp=False)

# Dự báo cho cột '7'
y_pred_7 = model_fit_7.predict(start=len(train_7), end=len(train_7) + len(test_7) - 1, dynamic=False)

# Vẽ kết quả cho cột '7'
plt.figure(figsize=(10, 6))
plt.plot(train_7.index, train_7, label='Train (Column 7)')
plt.plot(test_7.index, test_7, label='Test (Column 7)')
plt.plot(test_7.index, y_pred_7, label='SARIMA Prediction (Column 7)')
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('SARIMA Model for Column 7')
plt.legend()
plt.show()

from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Mean Absolute Error (MAE)
mae_6 = mean_absolute_error(test_6, y_pred_6)
mae_7 = mean_absolute_error(test_7, y_pred_7)

print('Mean Absolute Error (Column 6):', mae_6)
print('Mean Absolute Error (Column 7):', mae_7)

# Root Mean Squared Error (RMSE)
rmse_6 = mean_squared_error(test_6, y_pred_6, squared=False)
rmse_7 = mean_squared_error(test_7, y_pred_7, squared=False)

print('Root Mean Squared Error (Column 6):', rmse_6)
print('Root Mean Squared Error (Column 7):', rmse_7)

# Mean Absolute Percentage Error (MAPE)
def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

mape_6 = mean_absolute_percentage_error(test_6, y_pred_6)
mape_7 = mean_absolute_percentage_error(test_7, y_pred_7)

print('Mean Absolute Percentage Error (Column 6):', mape_6)
print('Mean Absolute Percentage Error (Column 7):', mape_7)

# R-squared (R2)
from sklearn.metrics import r2_score

r2_6 = r2_score(test_6, y_pred_6)
r2_7 = r2_score(test_7, y_pred_7)

print('R-squared (Column 6):', r2_6)
print('R-squared (Column 7):', r2_7)

```
