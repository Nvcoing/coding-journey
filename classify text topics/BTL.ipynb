```python
import pandas as pd

# Tạo DataFrame với dữ liệu bạn cung cấp
df = pd.read_excel('btl.xlsx', engine='openpyxl')

# Tạo DataFrame mới từ việc "stack" các cột và tạo thành các cột mới "Thông số" và "Nhãn"
df = pd.DataFrame(df[['Kinh tế','Thể thao','Công nghệ']].stack(), columns=['Văn bản'])
df['Nhãn'] = df.index.get_level_values(1)
df.index = range(len(df))

# Kết hợp ngẫu nhiên các hàng
df = df.sample(frac=1).reset_index(drop=True)

df.head(50)
```

```python
import pandas as pd

# df là DataFrame sau khi thực hiện one-hot encoding
df_encoded = pd.get_dummies(df['Nhãn'])

# Chuyển DataFrame thành mảng NumPy
matrix = df_encoded[['Công nghệ','Kinh tế','Thể thao']].values.astype(int)
matrix
```

```python
pd.concat([df['Nhãn'],
           pd.get_dummies(df['Nhãn']).astype(int)], axis=1).head(10)
```

```python
# Import các thư viện cần thiết
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from sklearn.model_selection import train_test_split
# Dữ liệu mẫu
texts = df['Văn bản']  # Lấy dữ liệu văn bản từ DataFrame df
labels = matrix  # Lấy nhãn từ ma trận nhãn đã được xác định trước
```

```python
# Tokenize văn bản
tokenizer = Tokenizer()  # Khởi tạo đối tượng Tokenizer
tokenizer.fit_on_texts(texts)  # Xây dựng từ điển từ văn bản
sequences = tokenizer.texts_to_sequences(texts)  # Chuyển đổi văn bản thành chuỗi số
sequences
```

```python
# Độ dài tối đa của mỗi chuỗi
max_sequence_length = max([len(seq) for seq in sequences])  # Tìm độ dài lớn nhất của các chuỗi số
# Điền chuỗi số với giá trị 0 để có cùng độ dài
data = pad_sequences(sequences, maxlen=max_sequence_length)
data
```

```python

# Số lượng từ vựng
vocab_size = len(tokenizer.word_index) + 1  # Số lượng từ vựng trong từ điển Tokenizer

# Xây dựng mô hình LSTM
model = Sequential([
    Embedding(vocab_size, 100, input_length=max_sequence_length),  # Lớp nhúng với kích thước từ vựng, kích thước vector nhúng là 100
    LSTM(100),  # Lớp LSTM với 100 đơn vị đầu ra
    Dense(3, activation='softmax')  # Lớp kết nối đầy đủ với 3 đầu ra và kích hoạt softmax
])

# Biên soạn mô hình
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
# Chia dữ liệu thành tập huấn luyện và tập kiểm tra (80% cho huấn luyện, 20% cho kiểm tra)
train_data, test_data, train_labels, test_labels = train_test_split(data, labels, test_size=0.2, random_state=42)

# Huấn luyện mô hình
model.fit(train_data, train_labels, epochs=5, batch_size=1)  # Sử dụng dữ liệu và nhãn để huấn luyện mô hình trong 10 epoch với kích thước batch là 1
# Đánh giá mô hình trên tập kiểm tra
loss, accuracy = model.evaluate(test_data, test_labels)
print("Test Accuracy:", accuracy)
from sklearn.metrics import precision_score, recall_score, f1_score

# Đánh giá mô hình trên tập kiểm tra
y_pred_prob = model.predict(test_data)
y_pred = np.argmax(y_pred_prob, axis=1)

precision = precision_score(np.argmax(test_labels, axis=1), y_pred, average='weighted')
recall = recall_score(np.argmax(test_labels, axis=1), y_pred, average='weighted')
f1 = f1_score(np.argmax(test_labels, axis=1), y_pred, average='weighted')

print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)

from sklearn.metrics import classification_report

report = classification_report(np.argmax(test_labels, axis=1), y_pred)
print("Báo cáo phân loại:\n", report)
```

```python

from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

# Tiền xử lý văn bản
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(texts)  # Chuyển đổi văn bản thành ma trận TF-IDF.Mỗi dòng là một văn bản

# Chuyển đổi ma trận nhãn từ one-hot encoding sang 1D array
labels_1d = np.argmax(labels, axis=1)

# Tạo và huấn luyện mô hình SVM

svm_model = SVC()
# Chia dữ liệu thành tập huấn luyện và tập kiểm tra (80-20 split)
X_train, X_test, y_train, y_test = train_test_split(X, labels_1d, test_size=0.2, random_state=42)

# Huấn luyện mô hình SVM trên tập huấn luyện
svm_model.fit(X_train, y_train)

# Đánh giá hiệu suất trên tập kiểm tra
accuracy = svm_model.score(X_test, y_test)
print("Độ chính xác trên tập kiểm tra:", accuracy)

# Dự đoán nhãn cho tập kiểm tra để xem các chỉ số khác nhau như precision, recall, f1-score
from sklearn.metrics import classification_report
y_pred = svm_model.predict(X_test)
report = classification_report(y_test, y_pred)
print("Báo cáo phân loại:\n", report)


# Dự đoán nhãn cho đoạn văn bản mới


new_text_vectorized = vectorizer.transform([new_text])
svm_prediction = svm_model.predict(new_text_vectorized)

# Dự đoán với SVM trả về chỉ số của nhãn, chúng ta cần chuyển đổi lại thành tên nhãn
predicted_label = ''
if svm_prediction[0] == 0:
    predicted_label = 'Công nghệ'
elif svm_prediction[0] == 1:
    predicted_label = 'Kinh tế'
elif svm_prediction[0] == 2:
    predicted_label = 'Thể thao'

print("Dự đoán (SVM):", predicted_label)
```

```python
from sklearn.linear_model import LogisticRegression

# Tạo và huấn luyện mô hình Logistic Regression
logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)

# Đánh giá hiệu suất trên tập kiểm tra
accuracy = logistic_model.score(X_test, y_test)
print("Độ chính xác trên tập kiểm tra (Logistic Regression):", accuracy)

# Dự đoán nhãn cho tập kiểm tra để xem các chỉ số khác nhau như precision, recall, f1-score
y_pred_logistic = logistic_model.predict(X_test)
report_logistic = classification_report(y_test, y_pred_logistic)
print("Báo cáo phân loại (Logistic Regression):\n", report_logistic)

# Dự đoán nhãn cho đoạn văn bản mới
logistic_prediction = logistic_model.predict(new_text_vectorized)

# Dự đoán với Logistic Regression trả về chỉ số của nhãn, chúng ta cần chuyển đổi lại thành tên nhãn
predicted_label_logistic = ''
if logistic_prediction[0] == 0:
    predicted_label_logistic = 'Công nghệ'
elif logistic_prediction[0] == 1:
    predicted_label_logistic = 'Kinh tế'
elif logistic_prediction[0] == 2:
    predicted_label_logistic = 'Thể thao'

print("Dự đoán (Logistic Regression):", predicted_label_logistic)
```
